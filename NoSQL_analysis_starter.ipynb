## Notebook Set Up

```python
# Import dependencies
# Module used to connect Python with MongoDB
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments_list = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
results = establishments_list.count_documents(query)
print(results)
# Display the first document in the results using pprint
first_document = establishments_list.find_one(query)
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
results = establishments_list.find(query)
result_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': "London"},'RatingValue':{'$gte':'4'}}
# Use count_documents to display the number of documents in the result
results = establishments_list.count_documents(query)
print(results)
# Display the first document in the results using pprint
first_document = establishments_list.find_one(query)
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
results = establishments_list.find(query)
result_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments_list.find_one({'BusinessName':'Penang Flavours'})["geocode"]["latitude"]
longitude = establishments_list.find_one({'BusinessName':'Penang Flavours'})["geocode"]["longitude"]

query = {'RatingValue': "5", "geocode": {"$within": {"$center": [[longitude,latitude],degree_search]}}}

# return only the BusinessName
fields = {"BusinessName": 1, "scores.Hygiene": 1, "geocode": 1, "_id": 0}

sort = [('scores.Hygiene', 1)]

# Limit the results to 5
limit = 5

# Capture the results to a variable
results = list(establishments_list.find(query,fields).sort(sort).limit(limit))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.json_normalize(results)
top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}
# Print the number of documents in the result
# Put the pipeline together
pipeline = [match_query, group_query, sort_values]
results = list(establishments_list.aggregate(pipeline))
print("Number of Local Authorities in result: ", len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
# Display the number of rows in the DataFrame
# Display the first 10 rows of the DataFrame
result_df = pd.DataFrame(results)
print("Rows in DataFrame: ", len(result_df))
result_df.head(10)
```
